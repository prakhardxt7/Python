```python
Q1=ipl.select_dtypes(include=np.number).quantile(0.25)  #Q1
Q3=ipl.select_dtypes(include=np.number).quantile(0.75)  #Q3
```

```python
#IQR
IQR=ipl.select_dtypes(include=np.number).quantile(0.75)-ipl.select_dtypes(include=np.number).quantile(0.25)

IQR
```

```python
#maximum and minimum

maximum = ipl.select_dtypes(include=np.number).max()
minimum = ipl.select_dtypes(include=np.number).max()
print(maximum)
print(minimum)
```

```python
Upper_Outlier = Q3+(1.5*IQR)
Lower_Outlier = Q1-(1.5*IQR)
print(Upper_Outlier)
print(Lower_Outlier)
```
